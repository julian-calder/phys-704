# Estimating $T_c$ from XY confidence plots

```python
from sympy.solvers import solve
from sympy import Symbol

from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np

def linear_func(x, A, B):
    return A + B*x

def cubic_func(x, A, B, C, D):
    return D*(x**3) + C*(x**2) + B*x + A

def calc_t(A1, B1, A2, B2):
    x = Symbol('x')
    return solve(((linear_func(x, A1, B1)) - linear_func(x, A2, B2)), x, rational=False)[0]
```

```python
LOWER_INDEX = TEMP_VALS.index(0.82)
UPPER_INDEX = TEMP_VALS.index(0.98)
N = UPPER_INDEX - LOWER_INDEX

x = np.array(TEMP_VALS[LOWER_INDEX:UPPER_INDEX])

T_c_vals = []
T_c_errs = []
# adding the -1 to do separate fitting for the final plot
for i in range(len(confidence_lower) - 1):
    y_lower = confidence_lower[i][LOWER_INDEX:UPPER_INDEX]
    y_lower_err = error_lower[i][LOWER_INDEX:UPPER_INDEX]

    param_lower, param_cov_lower = curve_fit(linear_func, x, y_lower, sigma = y_lower_err)

    perr_lower = np.sqrt(np.diag(param_cov_lower))

    A_val_lower = param_lower[0]
    A_err_lower = perr_lower[0]
    
    B_val_lower = param_lower[1]
    B_err_lower = perr_lower[1]
    
    y_err_lower = np.sqrt((1 / (N - 2)) * (np.sum((y_lower - linear_func(x, A_val_lower, B_val_lower)) ** 2)))


    y_upper = confidence_upper[i][LOWER_INDEX:UPPER_INDEX]
    y_upper_err = error_upper[i][LOWER_INDEX:UPPER_INDEX]
    
    param_upper, param_cov_upper = curve_fit(linear_func, x, y_upper, sigma = y_upper_err)
    
    perr_upper = np.sqrt(np.diag(param_cov_upper))
    
    A_val_upper = param_upper[0]
    A_err_upper = perr_upper[0]
    
    B_val_upper = param_upper[1]
    B_err_upper = perr_upper[1]
    
    y_err_upper = np.sqrt((1 / (N - 2)) * (np.sum((y_upper - linear_func(x, A_val_upper, B_val_upper)) ** 2)))

    T_c = calc_t(A_val_upper, B_val_upper, A_val_lower, B_val_lower)

    numerator_error = np.sqrt(A_err_upper ** 2 + A_err_lower ** 2)
    denominator_error = np.sqrt(B_err_lower ** 2 + B_err_upper ** 2)
    
    T_c_err = ((A_val_upper - A_val_lower) / (B_val_lower - B_val_upper)) * np.sqrt(((numerator_error ** 2) / (A_val_upper - A_val_lower)) ** 2 + ((denominator_error ** 2) / (B_val_lower - B_val_upper)) ** 2)

    print("T_c = " + str(T_c) + "\n" + "T_c error = " + str(T_c_err) + "\n")

    plt.plot(x, linear_func(x, A_val_lower, B_val_lower))
    plt.plot(x, confidence_lower[i][LOWER_INDEX:UPPER_INDEX])
    
    plt.plot(x, linear_func(x, A_val_upper, B_val_upper))
    plt.plot(x, confidence_upper[i][LOWER_INDEX:UPPER_INDEX])
    plt.axvline(x = T_c, color = 'k', label=f"$T_c = {T_c}$")
    plt.legend()
    plt.show()

    T_c_vals.append(T_c)
    T_c_errs.append(T_c_err)
```
