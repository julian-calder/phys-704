# Define function for calculating T_c from graph

```python
from sympy.solvers import solve
from sympy import Symbol

def calc_t(coeffs_above, coeffs_below):
    x = Symbol('x')
    return solve(((coeffs_below[0] * x**3) + (coeffs_below[1] * x**2) + (coeffs_below[2] * x) + coeffs_below[3]) - 
                 ((coeffs_above[0] * x**3) + (coeffs_above[1] * x**2) + (coeffs_above[2] * x) + coeffs_above[3]), x, rational=False)
```

## Extracting raw spin configuration data

```python
FILENAME = "xyvsT_full4x4"

TEMP_VALS = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.72, 0.74, 0.76, 0.78, 0.8, 0.82, 0.84, 0.86, 0.88, 0.9, 0.92, 0.94, 0.96, 0.98, 1.0, 1.05, 1.1, 1.15, 1.2, 1.25, 1.3, 1.35, 1.4, 1.45, 1.5, 1.55, 1.6, 1.65, 1.7, 1.75, 1.8] 
NUM_TEMPS = len(TEMP_VALS)
NUM_BINS = 5000
TRAIN_BINS = int(0.8 * NUM_BINS)
TEST_BINS = int(0.2 * NUM_BINS)
NUM_TRAIN = int(TRAIN_BINS * NUM_TEMPS)
NUM_TEST = int(TEST_BINS * NUM_TEMPS)

L = 4

import matplotlib.pyplot as plt
import numpy as np

file_string = "../data/spinconfigs/spinConfigs_" + FILENAME + ".txt"

data = np.loadtxt(file_string)

raw_spin_configs = np.zeros(shape=(NUM_TEMPS, NUM_BINS, L, L))

data_dict = {}

# initialize data dictionary
for temp in range(NUM_TEMPS):
    data_dict[TEMP_VALS[temp]] = []

# extract spin configs values 
for temp in range(NUM_TEMPS):
    for bin in range(NUM_BINS):            
        raw_spin_configs[temp][bin] = data[(temp * NUM_BINS) + bin].reshape(L, L)

# populate data dictionary with numpy array of temps
for temp in range(NUM_TEMPS):
    data_dict[TEMP_VALS[temp]] = raw_spin_configs[temp]

cos_dict = {}
sin_dict = {}

# extract sin and cos of spin configs
for temp in range(NUM_TEMPS):
    # initialize dictionary
    cos_dict[TEMP_VALS[temp]] = []
    sin_dict[TEMP_VALS[temp]] = []
    # populate dictionaries with sin values and cos values
    cos_dict[TEMP_VALS[temp]] = np.cos(data_dict[TEMP_VALS[temp]])
    sin_dict[TEMP_VALS[temp]] = np.sin(data_dict[TEMP_VALS[temp]])
```

## Populate and train on cos values

```python
# initialize arrays for training and testing images and labels
train_images = np.zeros(shape=(NUM_TRAIN, L, L))
shuffled_images = train_images

train_labels = np.zeros(NUM_TRAIN, dtype=int)
shuffled_labels = train_labels

test_images = np.zeros(shape=(NUM_TEST, L, L))
shuffled_test_images = test_images

test_labels = np.zeros(NUM_TEST, dtype=int)
shuffled_test_labels = test_labels


# populate train images and train labels with spin configs
offset = 0
# replacing data_dict with cos_dict
for temp, array in cos_dict.items():
    train_images[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = array[0:TRAIN_BINS]
    test_images[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = array[TRAIN_BINS:]
    
    # if lower than T_c, should be a 0
    if temp < 0.8816:
        train_labels[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = np.zeros(TRAIN_BINS)
        test_labels[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = np.zeros(TEST_BINS)
    else:
        train_labels[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = np.ones(TRAIN_BINS)
        test_labels[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = np.ones(TEST_BINS)
    
    offset += 1


def shuffle_pairs(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]

# shuffle our images and labels according to the same permutation
shuffled_images, shuffled_labels = shuffle_pairs(train_images, train_labels)

# shuffle test images and labels
shuffled_test_images, shuffled_test_labels = shuffle_pairs(test_images, test_labels)
```

```python
import tensorflow as tf

xy4x4_cos = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(L, L)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(2)
])

xy4x4_cos.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

xy4x4_cos.fit(shuffled_images, shuffled_labels, epochs=20)
```

```python
raw_confidence_above_vals = np.zeros(shape=(NUM_TEMPS, TEST_BINS))
raw_confidence_below_vals = np.zeros(shape=(NUM_TEMPS, TEST_BINS))

confidence_above_4x4cos = []
confidence_below_4x4cos = []
error_above_4x4cos = []
error_below_4x4cos = []

for temp in range(NUM_TEMPS):
    for bin in range(TEST_BINS):
        raw_confidence_above_vals[temp][bin] = float(predictions[(temp * TEST_BINS) + bin][1])
        raw_confidence_below_vals[temp][bin] = float(predictions[(temp * TEST_BINS) + bin][0])

    error_above_4x4cos.append(np.std(raw_confidence_above_vals[temp]) / np.sqrt(TEST_BINS))
    error_below_4x4cos.append(np.std(raw_confidence_below_vals[temp]) / np.sqrt(TEST_BINS))
    
    confidence_above_4x4cos.append(np.mean(raw_confidence_above_vals[temp]))
    confidence_below_4x4cos.append(np.mean(raw_confidence_below_vals[temp]))
```

```python
# make polynomial fit for confidence above T_c
poly_above_4x4_cos = np.polyfit(TEMP_VALS[6:40], confidence_above_4x4cos[6:40], 3)
# make polynomial fit for confidence below T_c
poly_below_4x4_cos = np.polyfit(TEMP_VALS[6:40], confidence_below_4x4cos[6:40], 3)

Tc_4x4_cos = calc_t(poly_above_4x4_cos, poly_below_4x4_cos)[1]

y1 = np.polyval(poly_above_4x4_cos, TEMP_VALS[6:40])
y2 = np.polyval(poly_below_4x4_cos, TEMP_VALS[6:40])
plt.plot(TEMP_VALS[6:40], confidence_above_4x4cos[6:40])
plt.plot(TEMP_VALS[6:40], y1)
plt.plot(TEMP_VALS[6:40], confidence_below_4x4cos[6:40])
plt.plot(TEMP_VALS[6:40], y2)
plt.axvline(x = Tc_4x4_cos, color = 'k', label=f"$T_c = {Tc_4x4_cos}$")
plt.legend()
plt.show()
```

```python
# initialize arrays for training and testing images and labels
# array will have an additional dimension to account for sin and cos together
train_images = np.zeros(shape=(NUM_TRAIN, L, L, 2))
shuffled_images = train_images

train_labels = np.zeros(NUM_TRAIN, dtype=int)
shuffled_labels = train_labels

test_images = np.zeros(shape=(NUM_TEST, L, L, 2))
shuffled_test_images = test_images

test_labels = np.zeros(NUM_TEST, dtype=int)
shuffled_test_labels = test_labels

offset = 0
for temp_offset in range(NUM_TEMPS):
    for bin in range(TRAIN_BINS):
        for row in range(L):
            for col in range(L):
                train_images[(temp_offset * TRAIN_BINS) + bin][row][col][0] = cos_dict[TEMP_VALS[temp_offset]][bin][row][col]
                train_images[(temp_offset * TRAIN_BINS) + bin][row][col][1] = sin_dict[TEMP_VALS[temp_offset]][bin][row][col]

    test_offset = 0
    for bin in range(TRAIN_BINS, NUM_BINS):
        for row in range(L):
            for col in range(L):
                test_images[(temp_offset * TEST_BINS) + test_offset][row][col][0] = cos_dict[TEMP_VALS[temp_offset]][bin][row][col]
                test_images[(temp_offset * TEST_BINS) + test_offset][row][col][1] = sin_dict[TEMP_VALS[temp_offset]][bin][row][col]

        test_offset += 1

    # if lower than T_c, should be a 0
    if TEMP_VALS[temp_offset] < 0.8816:
        train_labels[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = np.zeros(TRAIN_BINS)
        test_labels[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = np.zeros(TEST_BINS)
    else:
        train_labels[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = np.ones(TRAIN_BINS)
        test_labels[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = np.ones(TEST_BINS)
    
    offset += 1
    
# helper function to shuffle our images and labels
def shuffle_pairs(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]

# shuffle our images and labels according to the same permutation
shuffled_images, shuffled_labels = shuffle_pairs(train_images, train_labels)

# shuffle test images and labels
shuffled_test_images, shuffled_test_labels = shuffle_pairs(test_images, test_labels)
```

## Plot Magnetization squared as a function of time.

$$\bar{M_x} = \frac{1}{N}\sum{cos(\theta)}$$
$$\bar{M_y} = \frac{1}{N}\sum{sin(\theta)}$$

```python
x_magnetizations_raw_dict = {}
y_magnetizations_raw_dict = {}

x_magnetizations_raw = np.zeros(shape=(NUM_TEMPS, NUM_BINS))
y_magnetizations_raw = np.zeros(shape=(NUM_TEMPS, NUM_BINS))

for temp in range(NUM_TEMPS):
    for bin in range(NUM_BINS):
        # compute the sums 
        x_magnetizations_raw[temp][bin] = np.sum(cos_dict[TEMP_VALS[temp]][bin]) / (L**2)
        y_magnetizations_raw[temp][bin] = np.sum(sin_dict[TEMP_VALS[temp]][bin]) / (L**2)

    x_magnetizations_raw_dict[TEMP_VALS[temp]] = x_magnetizations_raw[temp]
    y_magnetizations_raw_dict[TEMP_VALS[temp]] = y_magnetizations_raw[temp]
```

## Load and preprocess data

```python
FILENAME = "xyvsT_full8x8"

TEMP_VALS = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.72, 0.74, 0.76, 0.78, 0.8, 0.82, 0.84, 0.86, 0.88, 0.9, 0.92, 0.94, 0.96, 0.98, 1.0, 1.05, 1.1, 1.15, 1.2, 1.25, 1.3, 1.35, 1.4, 1.45, 1.5, 1.55, 1.6, 1.65, 1.7, 1.75, 1.8] 
NUM_TEMPS = len(TEMP_VALS)
NUM_BINS = 5000
TRAIN_BINS = int(0.8 * NUM_BINS)
TEST_BINS = int(0.2 * NUM_BINS)
NUM_TRAIN = int(TRAIN_BINS * NUM_TEMPS)
NUM_TEST = int(TEST_BINS * NUM_TEMPS)

L = 8

import matplotlib.pyplot as plt
import numpy as np

file_string = "../data/spinconfigs/spinConfigs_" + FILENAME + ".txt"

data = np.loadtxt(file_string)

raw_spin_configs = np.zeros(shape=(NUM_TEMPS, NUM_BINS, L, L))

data_dict = {}

# initialize data dictionary
for temp in range(NUM_TEMPS):
    data_dict[TEMP_VALS[temp]] = []

# extract spin configs values 
for temp in range(NUM_TEMPS):
    for bin in range(NUM_BINS):            
        raw_spin_configs[temp][bin] = data[(temp * NUM_BINS) + bin].reshape(L, L)

# populate data dictionary with numpy array of temps
for temp in range(NUM_TEMPS):
    data_dict[TEMP_VALS[temp]] = raw_spin_configs[temp]

cos_dict = {}
sin_dict = {}

# extract sin and cos of spin configs
for temp in range(NUM_TEMPS):
    # initialize dictionary
    cos_dict[TEMP_VALS[temp]] = []
    sin_dict[TEMP_VALS[temp]] = []
    # populate dictionaries with sin values and cos values
    cos_dict[TEMP_VALS[temp]] = np.cos(data_dict[TEMP_VALS[temp]])
    sin_dict[TEMP_VALS[temp]] = np.sin(data_dict[TEMP_VALS[temp]])
```

```python
# initialize arrays for training and testing images and labels
train_images = np.zeros(shape=(NUM_TRAIN, L, L))
shuffled_images = train_images

train_labels = np.zeros(NUM_TRAIN, dtype=int)
shuffled_labels = train_labels

test_images = np.zeros(shape=(NUM_TEST, L, L))
shuffled_test_images = test_images

test_labels = np.zeros(NUM_TEST, dtype=int)
shuffled_test_labels = test_labels


# populate train images and train labels with spin configs
offset = 0
# replacing data_dict with cos_dict
for temp, array in cos_dict.items():
    train_images[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = array[0:TRAIN_BINS]
    test_images[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = array[TRAIN_BINS:]
    
    # if lower than T_c, should be a 0
    if temp < 0.8816:
        train_labels[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = np.zeros(TRAIN_BINS)
        test_labels[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = np.zeros(TEST_BINS)
    else:
        train_labels[(TRAIN_BINS * offset):(TRAIN_BINS * (offset+1))] = np.ones(TRAIN_BINS)
        test_labels[(TEST_BINS * offset):(TEST_BINS * (offset+1))] = np.ones(TEST_BINS)
    
    offset += 1


def shuffle_pairs(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]

# shuffle our images and labels according to the same permutation
shuffled_images, shuffled_labels = shuffle_pairs(train_images, train_labels)

# shuffle test images and labels
shuffled_test_images, shuffled_test_labels = shuffle_pairs(test_images, test_labels)
```

## Produce magnetization plot

```python
x_magnetizations_raw_dict = {}
y_magnetizations_raw_dict = {}

x_magnetizations_raw = np.zeros(shape=(NUM_TEMPS, NUM_BINS))
y_magnetizations_raw = np.zeros(shape=(NUM_TEMPS, NUM_BINS))

for temp in range(NUM_TEMPS):
    for bin in range(NUM_BINS):
        # compute the sums 
        x_magnetizations_raw[temp][bin] = np.sum(cos_dict[TEMP_VALS[temp]][bin]) / (L**2)
        y_magnetizations_raw[temp][bin] = np.sum(sin_dict[TEMP_VALS[temp]][bin]) / (L**2)

    x_magnetizations_raw_dict[TEMP_VALS[temp]] = x_magnetizations_raw[temp]
    y_magnetizations_raw_dict[TEMP_VALS[temp]] = y_magnetizations_raw[temp]

x_magnetizations_squared_dict = {}
y_magnetizations_squared_dict = {}

for temp in range(NUM_TEMPS):
    x_magnetizations_squared_dict[TEMP_VALS[temp]] = np.square(x_magnetizations_raw_dict[TEMP_VALS[temp]])
    y_magnetizations_squared_dict[TEMP_VALS[temp]] = np.square(y_magnetizations_raw_dict[TEMP_VALS[temp]])

avg_mx_squared_dict = {}
avg_my_squared_dict = {}

magnetization_errors = np.zeros(NUM_TEMPS)

for temp in range(NUM_TEMPS):
    avg_mx_squared_dict[TEMP_VALS[temp]] = np.mean(x_magnetizations_squared_dict[TEMP_VALS[temp]])
    avg_my_squared_dict[TEMP_VALS[temp]] = np.mean(y_magnetizations_squared_dict[TEMP_VALS[temp]])
    magnetization_errors[temp] = np.std(x_magnetizations_squared_dict[TEMP_VALS[temp]]) / np.sqrt(NUM_BINS)

avg_m_squared_dict = {}

for temp in TEMP_VALS:
    avg_m_squared_dict[temp] = avg_mx_squared_dict[temp] + avg_my_squared_dict[temp]

temp_vals = []
magnetization_vals = []

# iterate through dictionary of data to facilitate plotting
for temp, magnetization in avg_m_squared_dict.items():
    temp_vals.append(temp)
    magnetization_vals.append(magnetization)

magnetization_vals_64 = magnetization_vals
magnetization_errors_64 = magnetization_errors

plt.errorbar(temp_vals, magnetization_vals, magnetization_errors, marker=".", markersize=10)
plt.title("XY Model Magnetization vs. Temperature for L = 64")
plt.axvline(x = 0.8816, color = 'k', label="$T_{c} = 0.8816$")
plt.legend()
plt.xlabel("Temperature ($k_{b}T/J$)")
plt.ylabel("Average Magnetization Squared")
#plt.savefig('xy_magnetization_L=64.png', dpi=500)
plt.show()
```
