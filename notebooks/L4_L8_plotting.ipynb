# Parsing and plotting L = 4 data

```python
FILENAME = "demo4x4"
TEMP_VALS = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3.0, 3.25, 3.5, 3.75, 4.0, 4.25, 4.5, 4.75, 5.0]
NUM_TEMPS = 20
NUM_BINS = 10

import matplotlib.pyplot as plt
import numpy as np

file_string = "../bins_" + FILENAME + ".txt"

data = []

data_file = open(file_string, "r")
for row in data_file:
    # remove header rows so we are only working with numerical values
    if row[0] != ("#"): 
        row = row.split()
        # append 
        data.append(row)

# initialize dictionary to store temperature data and corresponding magnetizations
data_dict = {}

# populate keys for dictionary from temperature values
for val in TEMP_VALS:
    data_dict[val] = 0

# initialize list to store magnetization values
magnetization_vals = []

# extract magnetization values 
for row in data:
    magnetization_vals.append(float(row[-1]))

# collect each set of 10 magnetization measurements
for temp in range(NUM_TEMPS):
    magnetization = 0
    for bin in range(NUM_BINS):
        magnetization += magnetization_vals[(temp * NUM_BINS) + bin]
    # compute average magnetization
    magnetization /= NUM_BINS
    # store magnetization value associated with given temperature
    data_dict[TEMP_VALS[temp]] = magnetization

# initialize empty lists to store temperatures and average magnetizations
temps = []
magnetizations = []

# iterate through dictionary of data to facilitate plotting
for temp, magnetization in data_dict.items():
    temps.append(temp)
    magnetizations.append(magnetization)


plt.plot(temps, magnetizations)
plt.title("L = 4")
plt.xlabel("Temperature")
plt.ylabel("Average Magnetization")
plt.show()
```
