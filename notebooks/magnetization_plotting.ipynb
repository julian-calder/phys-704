# Parsing and plotting L = 4 data

```python
import matplotlib.pyplot as plt
import numpy as np

FILENAME = "demo4x4"
TEMP_VALS = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3.0, 3.25, 3.5, 3.75, 4.0, 4.25, 4.5, 4.75, 5.0]
NUM_TEMPS = len(TEMP_VALS)
NUM_BINS = 10

file_string = "../bins_" + FILENAME + ".txt"

data = np.loadtxt(file_string)

raw_magnetization_vals = np.zeros(shape=(NUM_TEMPS, NUM_BINS))
magnetization_errors = np.zeros(NUM_TEMPS)

data_dict = {}

# extract magnetization values 
for temp in range(NUM_TEMPS):
    for bin in range(NUM_BINS):
        raw_magnetization_vals[temp][bin] = float(data[(temp * NUM_BINS) + bin][-1])

    data_dict[TEMP_VALS[temp]] = np.mean(raw_magnetization_vals[temp])
    magnetization_errors[temp] = np.std(raw_magnetization_vals[temp]) / np.sqrt(NUM_BINS)

temps_4x4 = []
magnetizations_4x4 = []

# iterate through dictionary of data to facilitate plotting
for temp, magnetization in data_dict.items():
    temps_4x4.append(temp)
    magnetizations_4x4.append(magnetization)

plt.errorbar(temps_4x4, magnetizations_4x4, magnetization_errors, marker=".", markersize=15, color="k")
plt.title("L = 4 with error bars")
plt.xlabel("Temperature")
plt.ylabel("Average Magnetization")
plt.show()
```

```python
plt.plot(temps_4x4, magnetization_errors)
plt.title("L = 4 with error bars")
plt.xlabel("Temperature")
plt.ylabel("Error")
plt.show()
```

# Parsing and plotting L = 8 data for temp ranges from 0.25 to 5 (interval of 0.25)

```python
FILENAME = "demo8x8"
TEMP_VALS = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3.0, 3.25, 3.5, 3.75, 4.0, 4.25, 4.5, 4.75, 5.0]
NUM_TEMPS = len(TEMP_VALS)
NUM_BINS = 10

import matplotlib.pyplot as plt
import numpy as np

file_string = "../bins_" + FILENAME + ".txt"

data = []

data_file = open(file_string, "r")
for row in data_file:
    # remove header rows so we are only working with numerical values
    if row[0] != ("#"): 
        row = row.split()
        # append 
        data.append(row)

# initialize dictionary to store temperature data and corresponding magnetizations
data_dict = {}

# populate keys for dictionary from temperature values
for val in TEMP_VALS:
    data_dict[val] = 0

# initialize numpy array to store magnetization values
raw_magnetization_vals = np.zeros(shape=(NUM_TEMPS, NUM_BINS))

magnetization_errors = np.zeros(NUM_TEMPS)

# extract magnetization values 
for temp in range(NUM_TEMPS):
    for bin in range(NUM_BINS):
        raw_magnetization_vals[temp][bin] = float(data[(temp * NUM_BINS) + bin][-1])

    data_dict[TEMP_VALS[temp]] = np.mean(raw_magnetization_vals[temp])
    magnetization_errors[temp] = np.std(raw_magnetization_vals[temp]) / np.sqrt(NUM_BINS)

print(magnetization_errors)

temps_8x8 = []
magnetizations_8x8 = []

# iterate through dictionary of data to facilitate plotting
for temp, magnetization in data_dict.items():
    temps_8x8.append(temp)
    magnetizations_8x8.append(magnetization)

magnetizations_array_8x8 = np.array(magnetizations_8x8)

plt.errorbar(temps_8x8, magnetizations_8x8, magnetization_errors)
plt.title("L = 8 with error bars")
plt.xlabel("Temperature")
plt.ylabel("Average Magnetization")
plt.show()
```
